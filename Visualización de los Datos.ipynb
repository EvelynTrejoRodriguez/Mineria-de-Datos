Un scatter plot o tambiÃ©n llamado grÃ¡fico de dispersiÃ³n se utiliza para trazar puntos de datos en los ejes horizontales y verticales con el objetivo de mostrar cuÃ¡nto afecta una variable a otra. Cada punto representa una observaciÃ³n. La posiciÃ³n en el eje X (horizontal) e Y (vertical) representa los valores de las 2 variables. TambiÃ©n es posible mapear otra variable al tamaÃ±o de cada punto, lo que hace que una burbuja se dibuje. Si tienes muchos puntos y luchas con la sobretrazaciÃ³n, considera el uso de la grÃ¡fica de densidad 2D.

A comparaciÃ³n de un grÃ¡fico de hexbin, que es Ãºtil para representar la relaciÃ³n de 2 variables numÃ©ricas cuando se tienen muchos puntos de datos. En lugar de superponerse, como en el scatter plot, la ventana de trazado se divide en varios hexÃ¡gonos y se cuenta el nÃºmero de puntos por hexÃ¡gono. El color denota este nÃºmero de puntos, tomando en cuenta que podemos cambiar el tamaÃ±o de los bins usando el argumento gridsize.

Podemos utilizar tanto un grÃ¡fico de dispersiÃ³n como un hexbin, la diferencia aquÃ­ es que mientras que los grÃ¡ficos de dispersiÃ³n muestran cada punto, los grÃ¡ficos de hexÃ¡gono determinan el color de cada hexÃ¡gono tomando la media de todos los valores de los precios dentro de ese hexÃ¡gono.

Es Ã³ptimo el grÃ¡fico hexbin para cuando tenemos una mayor cantidad de datos, siendo mÃ¡s fÃ¡cil de analizar al observarlo, a comparaciÃ³n del grÃ¡fico de dispersiÃ³n siendo optimo para una cantidad menor de datos, ya que tambiÃ©n su grÃ¡fico muestra esa relaciÃ³n, pero al ser muchos datos, se sobreponen encima de, perdiendo informaciÃ³n.
