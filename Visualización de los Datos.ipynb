Un scatter plot o también llamado gráfico de dispersión se utiliza para trazar puntos de datos en los ejes horizontales y verticales con el objetivo de mostrar cuánto afecta una variable a otra. Cada punto representa una observación. La posición en el eje X (horizontal) e Y (vertical) representa los valores de las 2 variables. También es posible mapear otra variable al tamaño de cada punto, lo que hace que una burbuja se dibuje. Si tienes muchos puntos y luchas con la sobretrazación, considera el uso de la gráfica de densidad 2D.

A comparación de un gráfico de hexbin, que es útil para representar la relación de 2 variables numéricas cuando se tienen muchos puntos de datos. En lugar de superponerse, como en el scatter plot, la ventana de trazado se divide en varios hexágonos y se cuenta el número de puntos por hexágono. El color denota este número de puntos, tomando en cuenta que podemos cambiar el tamaño de los bins usando el argumento gridsize.

Podemos utilizar tanto un gráfico de dispersión como un hexbin, la diferencia aquí es que mientras que los gráficos de dispersión muestran cada punto, los gráficos de hexágono determinan el color de cada hexágono tomando la media de todos los valores de los precios dentro de ese hexágono.

Es óptimo el gráfico hexbin para cuando tenemos una mayor cantidad de datos, siendo más fácil de analizar al observarlo, a comparación del gráfico de dispersión siendo optimo para una cantidad menor de datos, ya que también su gráfico muestra esa relación, pero al ser muchos datos, se sobreponen encima de, perdiendo información.
