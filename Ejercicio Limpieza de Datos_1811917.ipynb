### PREGUNTAS DE INTERÉS:

#### ¿Para que me puede servir estas agrupaciones para la práctica siguiente?

##### Al realizar archivos de cada agrupación, nos facilitará el manejo de la información, ya que estarán clasificadas por día o distrito en donde se encuentra el Departamento de Policía que tomo el crimen, siendo relevante para nosotros debido a que gracias a esto, nos ayudará a resolver nuestra problemática principal.

```python
newbase['Resolucion'] = newbase['Resolucion'].str.replace(',','_')
newbase['Resolucion'] = newbase['Resolucion'].str.replace(' ','_')
newbase['Resolucion'] = newbase['Resolucion'].str.replace('-','_')
```

```python
newbase.Resolucion.replace(to_replace = dict(NONE= 0,UNFOUNDED=0,NOT_PROSECUTED=0,COMPLAINANT_REFUSES_TO_PROSECUTE=0), inplace = True)
newbase.Resolucion.replace(to_replace = dict(ARREST__BOOKED = 1, EXCEPTIONAL_CLEARANCE = 1, JUVENILE_BOOKED=1,ARREST__CITED=1,PSYCHOPATHIC_CASE=1,LOCATED=1,CLEARED_CONTACT_JUVENILE_FOR_MORE_INFO=1,JUVENILE_DIVERTED=1,PROSECUTED_BY_OUTSIDE_AGENCY=1,JUVENILE_CITED=1), inplace = True)
```
